```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

with tf.device('/cpu:0'):
    
    x = tf.placeholder(tf.float32, shape=(), name="X")
    y = tf.placeholder(tf.float32, shape=(), name="Y")
        
    w = tf.Variable(5, dtype=tf.float32, name='W')
    v = tf.Variable(10, dtype=tf.float32, name='V')
    
    loss = (x * w - y * v) ** 2
    
    optimizer = tf.train.GradientDescentOptimizer(1e-4)
    objective = optimizer.minimize(loss)
    
    
    
```

```python
init_op = tf.global_variables_initializer()

with tf.Session() as sess:
    
    sess.run(init_op)
    
    for i in range(100):
        _, loss_value, w_value, v_value = sess.run([objective, loss, w, v], feed_dict={x : 15, y : -5})
        print(loss_value, w_value, v_value)
    
    
    
    
```
